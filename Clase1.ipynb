```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
print(iris['DESCR'])
```

```python
X_iris = iris.data
y_iris = iris.target
```

```python
X_iris.shape
```

```python
y_iris.shape
```

```python
Xtr_iri, Xte_iri, ytr_iri, yte_iri = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42, stratify=y_iris)
```

```python
scaler_i = StandardScaler()
Xtr_iri_s = scaler_i.fit_transform(Xtr_iri)
Xte_iri_s = scaler_i.transform(Xte_iri)
```

```python
log_iris = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500)
log_iris.fit(Xtr_iri_s, ytr_iri)
```

```python
Xtr_iri_s.shape
```

```python
yhat_iri = log_iris.predict(Xte_iri_s)
```

```python
yhat_iri
```

```python
print('--- Iris: Reporte de clasificación (Multiclase) ---')
print(classification_report(yte_iri, yhat_iri, target_names=iris.target_names, digits=4))
```

```python
cm_iri = confusion_matrix(yte_iri, yhat_iri)
sns.heatmap(cm_iri, annot=True, fmt='d', cmap='Greens')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.title('Matriz de Confusión - Iris (Logistic Regression Multinomial)')
plt.show()
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(np.vstack([Xtr_iri_s, Xte_iri_s]))
Xtr_red = X_reduced[:len(Xtr_iri_s)]
Xte_red = X_reduced[len(Xtr_iri_s):]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Xtr_red[:, 0], Xtr_red[:, 1], c=ytr_iri, cmap='viridis',
            alpha=0.6, edgecolors='k', s=50, label='Training')
plt.scatter(Xte_red[:, 0], Xte_red[:, 1], c=yte_iri, cmap='viridis',
            alpha=0.8, marker='s', edgecolors='k', s=50, label='Test')

plt.xlabel('Primera Componente Principal (PC1)')
plt.ylabel('Segunda Componente Principal (PC2)')
plt.title('Visualización de Datos Iris con PCA (2D)')
plt.colorbar(label='Clase')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
log_iris_2d = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500)
log_iris_2d.fit(Xtr_red, ytr_iri)
```

```python
Xtr_red.shape
```

```python
yhat_iri_2d = log_iris_2d.predict(Xte_red)
```

```python
print('--- Iris: Reporte de clasificación (Multiclase) 2 dimensiones ---')
print(classification_report(yte_iri, yhat_iri_2d, target_names=iris.target_names, digits=4))
```

```python
# Grilla para la frontera
xx, yy = np.meshgrid(np.linspace(X_reduced[:,0].min()-1, X_reduced[:,0].max()+1, 200),
                     np.linspace(X_reduced[:,1].min()-1, X_reduced[:,1].max()+1, 200))
Z = log_iris_2d.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.2)
plt.scatter(Xtr_red[:,0], Xtr_red[:,1], c=ytr_iri, cmap='viridis', edgecolor='k', label='Train')
plt.scatter(Xte_red[:,0], Xte_red[:,1], c=yte_iri, cmap='viridis', marker='x', s=80, label='Test')
plt.title('Fronteras de decisión (Iris) - PCA 2D + Logistic Regression')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    n_samples = 200,
    n_features = 2,
    n_redundant = 0,
    n_clusters_per_class = 1,
    n_classes = 3,
    random_state = 42
)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_scaled, y)
```

```python
h = 0.02  # tamaño del paso
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# --- 5. Predecir sobre la malla ---
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# --- 6. Graficar ---
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Set1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, edgecolor='k', cmap=plt.cm.Set1)
plt.title("Frontera de decisión - k-NN (k=1)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
log = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=500)
log.fit(X_train_s, y_train)
```

```python
y_pred_log = log.predict(X_test_s)
```

```python
print(classification_report(y_test, y_pred_log))
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_s, y_train)
```

```python
y_pred_knn = knn.predict(X_test_s)
```

```python
print(classification_report(y_test, y_pred_knn))
```

```python
nb = GaussianNB()
nb.fit(X_train_s, y_train)
```

```python
y_pred_nb = nb.predict(X_test_s)
```

```python
print(classification_report(y_test, y_pred_nb))
```

```python
cm_nb = confusion_matrix(y_test, y_pred_nb)
sns.heatmap(cm_nb, annot=False, cmap='Blues')
plt.title('Matriz de Confusión – Naive Bayes (Digits)')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()
```
